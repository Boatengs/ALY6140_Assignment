## Quetion 8
###  Name the top 10 champions who have won highest championship titles - display champion name, country, 

```python
# Your solution here
# HINT: use pandas groupby() function
```

```python
import pandas as pd

# Reading data .csv as a pandas df
df = pd.read_csv('wimbledons_champions.csv')

# Groupn the df by Champion and counting the num of wins
champion_wins = df.groupby(['Champion', 'Champion Nationality']).size().reset_index(name='Wins')

# Sortnthe df by num of wins in descending order
sorted_champion_wins = champion_wins.sort_values(by='Wins', ascending=False)

# Selectn the top 10 champs with the highest numb of wins
top_10_champions = sorted_champion_wins.head(10)

# Displn the top 10 champs with their country and num of wins
print(top_10_champions.to_string(index=False))
```
